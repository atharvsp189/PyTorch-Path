### Creating Tensors

```python
# using empty
# takes shape of a tensors as arguments
a = torch.empty(2, 3)
print(a)
print(f"Type: {type(a)}")
# just assigns memory - no assignments
```

```python
# using zeros
# assign zeros to the value of tensors
# takes shape of a tensors as arguments
torch.zeros(2, 3)
```

```python
# using ones
# assign one to the value of tensors
# takes shape of a tensors as arguments
torch.ones(2, 3)
```

```python
# using rand
# assign random values to the value of tensors
# takes shape of a tensors as arguments

# add seed for reproducibility
torch.manual_seed(42)
torch.rand(2, 3)
```

```python
# using tensor
# takes tensor as an argument
torch.tensor([[1,2], [3,4]])
```

```python
# arange - assign serial values
# takes length as an arguments
torch.arange(10)
```

```python
# using linspace
# linearly range
# argument - start_range, end_range, evenly_spread
# 0, 10, 10 -> 10 evenly spread values between 0 to 10
# 0, 10, 20 -> 20 evenly spread values between 0 to 10
a = torch.linspace(0, 10, 10) 
print("A: ", a)
b = torch.linspace(0, 10, 20) 
print("B: ", b)
```

```python
# using eye
# Identity matrix - diagonal element 1 else 0
torch.eye(5)
```

```python
# using full
# takes shape and values as an argument
# assign value to every place
torch.full((2, 3, 2), 5)
```

### Mathematical Operations

```python
x = torch.rand(2, 3, dtype=torch.float16)
x
```

#### scaler operations

```python
# addition
print(x + 3)
# subtraction
print(x - 3)
# multiplication
print(x * 3)
# division
print(x / 3)
# mod
print(x % 3)
# power
print(x ** 3)
```

```python
# element wise operation

a = torch.rand(2, 3)
b = torch.rand(2, 3)
```

```python
# addition
print(a + b)
# subtraction
print(a - b)
# multiplication
print(a * b)
# division
print(a / b)
# mod
print(a % b)
# power
print(a ** b)
# abs
print(torch.abs(a))
# neg
print(torch.neg(a))
# round
print(torch.round(a))
# ceil
print(torch.ceil(a))
# floor
print(torch.floor(a))
# clamp
# convert tensor to a certain range
print(torch.clamp(a, min=2, max=3))
```

#### Reduction Operation

```python
# reduce tensor to a single number

print("Sum :")
# sum
print(torch.sum(a))

# sum along columns
print(torch.sum(a, dim=0))

# sum along rows
print(torch.sum(a, dim=1))

print("Mean :")
# mean
print(torch.mean(a))

# mean along columns
print(torch.mean(a, dim=0))

# mean along rows
print(torch.mean(a, dim=1))
```

```python
# Similar to this we have

# Median -> median
# Std -> std
# Variance -> var
# Product -> prod
# min -> min
# max -> max
# min position -> argmin
# max position -> argmax
```

#### Matrix Operations

```python
a = torch.rand(2, 2)
b = torch.rand(2, 2)
```

```python
# matrix multiplication

print(torch.matmul(a, b))
```

```python
# Transpose
# 0 -> first_dim 1 -> second_dim
print(torch.transpose(a, 0, 1))
```

```python
# Determinant
print(torch.det(a))
```

```python
# Inverse
print(torch.inverse(a))
```

#### Comparision operation

```python
a = torch.rand(2, 2)
b = torch.rand(2, 2)
```

```python
print(a < b)
print(a > b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

#### Special Function - logs, sqrt, sigmoid

```python
a = torch.rand(2, 3)
```

```python
print(torch.log(a))
print(torch.sqrt(a))
print(torch.sigmoid(a))
print(torch.relu(a))
```

### Inplace Operations

```python
# stores the new result in that place only
# _ reprent inplace operatiosn
a = torch.rand(2, 3)
print(a)
b = torch.rand(2, 3)
print(b)
```

```python
a.add_(b)
a
```

```python
torch.relu_(a)
a
```

#### Copying Tensor

```python
# Copying the tensor


# = -> creates shallow copy

# for deep copy - use clone functio.
```

```python
a = torch.rand(2, 2)
print(id(a))
```
