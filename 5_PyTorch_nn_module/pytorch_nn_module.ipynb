```python
# creating sequential container

# create model class
class SimpleNN(nn.Module): # inheriting from nn module
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__() # initialize the parent class
        
        # sequential container
        self.network = nn.Sequential(
            nn.Linear(input_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, output_size),
            nn.Sigmoid()
        )
        

    # forward pass
    def forward(self, x):
        out = self.network(x)
        return out
```

Creating Synthetic Dataset

```python
features = torch.randn((10, 5))
```

Creating Model Instance

```python
model = SimpleNN(input_size=5, hidden_size=3, output_size=1)
```

```python
# internally override __call__ method
# will automatically call forward method
model(features)
```

Inspecting Model

```python
# to get the weights and biases
# <instance>.<layer_name>.<weight or bias>
print("Layer 1")
print(f"Weights: {model.l1.weight}")
print(f"Bias: {model.l1.bias}")   
print("Layer 2")
print(f" Weights: {model.l2.weight}")
print(f" Bias: {model.l2.bias}")   
```
